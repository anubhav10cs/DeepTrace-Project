**What is a language model/ NLP?**

Answer- So basically it is a deep learning model which recognizes relationship between text data and tries to find some patterns etc through neural networks/ transformers to predict the next word. Its primary objective is giving it a piece of text and for it to predict the next word after it, something like a autofill/suggestion option in various web browsers.

***Steps involved in language models-***

**1.Tokenization**

So the model cannot really understand words, it can only understand numbers, so each word is broken down, as it cannot understand relationship from looking at the entire sentence all together, but it can learn relationships from small words. So the process of breaking down a sentence in small parts, usually words or into smaller units(syllabals) is called Tokenization. 

Example-
Input Sentence - I am Anubhav Chatterjee

Tokens = ["I", "am" , "Anubhav" ,"Chatterjee"]

However it is not neccesary that it always converts it into words it can also break it down other wise , example-

Input Sentence - Unbelievable

Tokens = ["Un", "believable"]

**2.TokenId**

Each of the tokens produced are given a token id, which is at random and are given from a lookup table, which contains all the given vocabulary of the model.

**3. Embedding**

Each of these tokenId are converted into a vectors of some dimensions, these vector helps us to understanding relationship between the words after the model is trained. Initially it is also assigned randomly from a lookup table, but after the model is trained it changes and corrctly establishes relationship between words.
If two words have similar meaning they usually have similar embeddings.




**4. Self attention**

Now comes the main part , it is for the model to understand what word mean what in which context, for example the word bank can mean a river bank or the money bank, so our model should understand what kind of bank it is. To understand this  it uses something called query,key,value. Each word has a query which is basically asking a question like suppose the word bank might be asking thr question what is the context of using thse words. Then each query is compared with key values from previous words usually and a query key attention score, then it is normalized using softmax. After that we use attention score and do a weighted multiplication with the value matrix, which tells us the actual content.
